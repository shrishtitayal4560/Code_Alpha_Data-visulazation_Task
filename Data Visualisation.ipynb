```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def con(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(con)
```

# converted 0 & 1 value of senoir citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb["Churn"], labels = gb.index,autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'gender', data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'SeniorCitizen', data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# 2. Calculate row-wise percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# 3. Plot
fig, ax = plt.subplots(figsize=(4,4))

bottom = [0] * len(ct_percent)
labels = ct.columns

for churn_val in labels:
    values = ct_percent[churn_val]
    ax.bar(ct_percent.index, values, bottom=bottom, label=churn_val)
    
    # Add percentage labels on each bar segment
    for i, (val, btm) in enumerate(zip(values, bottom)):
        ax.text(i, btm + val / 2, f'{val:.1f}%', ha='center', va='center', color='white', fontsize=10)
    
    # Update bottom for stacking
    bottom = [btm + val for btm, val in zip(bottom, values)]

# Customize plot
ax.set_xticks([0, 1])
ax.set_xticklabels(['Not Senior', 'Senior'])
ax.grid(False)  # Disable grid
plt.title("Churn by Senior Citizen (in %)")
plt.ylabel("Percentage")
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 50, hue = "Churn")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set plot style
sns.set(style="whitegrid")

# Create subplots - 3 rows, 3 columns (3x3 = 9 plots)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each countplot
for i, col in enumerate(cols):
    ax = axes[i]
    sns.countplot(data=df, x=col, ax=ax, palette='Set2', hue = df["Churn"])
    ax.set_title(f'Count of {col}')
    ax.tick_params(axis='x', rotation=20)  # Rotate x-axis labels if needed
    ax.bar_label(ax.containers[0])
    ax.grid(False)  # Disable grid

# Hide any unused subplots (only needed if number of columns < total subplots)
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
